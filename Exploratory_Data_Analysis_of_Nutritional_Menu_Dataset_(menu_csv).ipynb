**Each Category like at
point in time**

```python
# Step 1: Group by Category
grouped = df.groupby('Category')[['Calories', 'Total Fat', 'Protein', 'Sugars']].mean().reset_index()

# Step 2: Plot category-based trends
plt.figure(figsize=(12,6))
plt.plot(grouped['Category'], grouped['Calories'], marker='o', label='Calories')
plt.plot(grouped['Category'], grouped['Total Fat'], marker='o', label='Total Fat')
plt.plot(grouped['Category'], grouped['Protein'], marker='o', label='Protein')
plt.plot(grouped['Category'], grouped['Sugars'], marker='o', label='Sugars')

plt.title("Average Nutrition by Food Category")
plt.xlabel("Food Category")
plt.ylabel("Average Nutrient Value")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

**Analyzing Product Distribution and customer**

```python
# Count of menu items per category
df['Category'].value_counts()
```

```python
#product nutrition analaysis
# Average nutrition by food category
category_stats = df.groupby('Category')[['Calories', 'Total Fat', 'Sugars', 'Protein']].mean().sort_values(by='Calories', ascending=False)
category_stats
```

**Visualization (Product Comparison)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot average nutrients per category
category_stats.plot(kind='bar', figsize=(12,6))
plt.title("Average Nutritional Values by Category")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

**1. Bar Chart – Number of Items per Category**

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index, palette='viridis')
plt.title("Number of Menu Items per Category")
plt.xlabel("Item Count")
plt.ylabel("Category")
plt.grid(True)
plt.show()
```

**2. Line Plot – Average Nutrients by Category**

```python
# Group averages
category_avg = df.groupby('Category')[['Calories', 'Total Fat', 'Protein', 'Sugars']].mean().reset_index()

# Plot lineplot
plt.figure(figsize=(12,6))
sns.lineplot(data=category_avg.set_index('Category'), markers=True)
plt.title("Average Nutritional Value by Category")
plt.ylabel("Average Value")
plt.xlabel("Food Category")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

** 3. Line Plot – Simulated Time Series of Calories Over Items**

```python
# Simulate time as item order
df['Time'] = range(len(df))

plt.figure(figsize=(12,5))
sns.lineplot(data=df, x='Time', y='Calories', label='Calories', color='darkorange')
plt.title("Simulated Time Series: Calories Over Menu Items")
plt.xlabel("Item Order (Simulated Time)")
plt.ylabel("Calories")
plt.grid(True)
plt.show()
```

**4.Heatmaps**


```python
# Create heatmap of normalized values
heatmap_data = df.groupby('Category')[['Calories', 'Total Fat', 'Protein', 'Sugars']].mean()

# Normalize the values for better color scaling
heatmap_normalized = (heatmap_data - heatmap_data.min()) / (heatmap_data.max() - heatmap_data.min())

plt.figure(figsize=(10,6))
sns.heatmap(heatmap_normalized, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Normalized Nutrient Heatmap by Category")
plt.show()
```

**Final Recommendations**
## 📌 Final Recommendations Based on EDA of menu.csv

Based on the exploratory data analysis performed on the `menu.csv` file, the following recommendations are drawn from observed trends, descriptive statistics, and visualizations:

1. **Reduce Calories in High-Calorie Categories**  
   The analysis showed that **Breakfast** and **Beef** items have the highest average calorie content. Introducing lighter options (e.g., smaller portions, low-fat alternatives) in these categories could support customers seeking healthier choices.

2. **Offer Low-Sugar Alternatives in Drinks and Desserts**  
   The **Drinks** and **Desserts** categories contain extremely high sugar content compared to other items. Offering sugar-free beverages, natural sweeteners, or fruit-based desserts would enhance menu balance and appeal to health-conscious consumers.

3. **Promote Protein-Rich Items in Chicken and Beef**  
   The **Chicken** and **Beef** categories were observed to have the highest protein values. These can be marketed as "High-Protein Power Meals" for gym-goers or those focusing on muscle recovery and strength.

4. **Reduce Sodium Levels in Sandwiches and Breakfast Items**  
   Visualizations and averages revealed **high sodium** levels in **Sandwiches** and **Breakfast** categories. Reducing processed ingredients and offering low-sodium variants would be beneficial.

5. **Add More Healthy and Vegetarian Options**  
   The dataset showed limited or no representation of **salads**, **vegetarian**, or **fruit-based items**. Expanding these categories would provide a more diverse and health-aligned menu for a broader customer base.
