```python
import pandas as pd

# Load the Netflix dataset (replace with your file path if needed)
df = pd.read_csv("netflix_titles.csv")

# Display the first 5 rows
df.head(5)
```

```python
print("Shape of dataset:", df.shape)  # (rows, columns)
```

```python
# Summary of missing values
print("Missing values per column:\n", df.isnull().sum())

# Optional: Check for inconsistent data types
print("\nData types:\n", df.dtypes)

# Optional: Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Check for unique values in the 'rating' column (could contain unexpected entries)
print(df['rating'].unique())

# Check for unique values in the 'country' column (could contain inconsistent data)
print(df['country'].unique())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check the number of rows after removing duplicates
print(df.shape)
```
